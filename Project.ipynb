```python
#------------------------------------- IMPORTS --------------------------------------------

import nltk
#nltk.download('punkt')
#nltk.download('stopwords')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import string
import re
import contractions
import matplotlib.pyplot as plt
import os
```

```python
#tokenize
def tokenize(text):
    expanded_words = []
    for word in text.split():
        expanded_words.append(word)
        
    return expanded_words
```

```python
def removeContractions(words):
    expanded_words = []
    for word in words:
        expanded_words.append(contractions.fix(word))
        
    expanded_text = ' '.join(expanded_words)
    return expanded_text
```

```python
def remove_stop(words):
    stop_words = set(stopwords.words("english"))
    stopRemovedFirst = [word for word in words if not word in stop_words]
    string = ' '.join(stopRemovedFirst)
    return string
```

```python
# Remove Stop Words
def removeALLStopWords(tokenized):
    stopRemovedOne = remove_stop(tokenized)
    stopRemovedOneTokenized = tokenize(stopRemovedOne)
    contractionsRemoved = removeContractions(stopRemovedOneTokenized)
    stopRemovedTwo = remove_stop(word_tokenize(contractionsRemoved))
    return stopRemovedTwo
```

```python
#Remove Punct
def removePunct(text):
    punct = string.punctuation
    textWithoutPunct = text
    for c in text:
        if c in punct:
            textWithoutPunct = textWithoutPunct.replace(c, "")
    
    textWithoutPunct = textWithoutPunct.replace(" s "," ")
    return textWithoutPunct
```

```python
#Remove Space
def removeExtraSpaces(text):
    output = " ".join(text.split())
    return output
```

```python
#finding unique
def uniqueWords(words):
    #finding unique
    unique = []
    for word in words:
        if word not in unique:
            unique.append(word)
            
    return unique
```

```python
def genVADDic():
    dictionary = {}
    vad = open("/home/ubuntu/Semester10/NLP/Project/NRC-VAD/NRC-VAD-Lexicon.txt")
    vad = vad.read()
    vadSplitNewLine = vad.split("\n")
    #print(vadSplitNewLine)
    
    for line in vadSplitNewLine:
        lineNoTabs = line.split("\t")
        VADValue = lineNoTabs[1:]
        dictionary[lineNoTabs[0]] = VADValue
    
    return dictionary
    #print(dictionary)
```

```python
def matchScriptToVAD(tokenizedText, Dictionary):
    
    ValenceVector = []
    ArousalVector = []
    DominanceVector = []
    
    for word in tokenizedText:
        if word in Dictionary.keys():
            valence = float(Dictionary[word][0])
            arousal = float(Dictionary[word][1])
            dominance = float(Dictionary[word][2])
            
            ValenceVector.append(valence)
            ArousalVector.append(arousal)
            DominanceVector.append(dominance)
            
        else:
            ValenceVector.append(0)
            ArousalVector.append(0)
            DominanceVector.append(0)
            
    return ValenceVector, ArousalVector, DominanceVector
```

```python
def plotVAD(avV, avA, avD, scriptName):

    scriptName = scriptName.split("/")[-1]
    plt.plot(avV, label= "Valence")
    plt.plot(avA, label= "Arousal")
    plt.plot(avD, label= "Dominance")
    plt.legend()
    scriptName = scriptName[:len(scriptName)-4]
    directory = "/home/ubuntu/Semester10/NLP/Project/VAD Plots/"+scriptName+".jpg"
    plt.title(scriptName)
    plt.savefig(directory)
    plt.close()
    #plt.show()
```

```python
def FindVAD():
    
    
    # ---------------------------------- Read Data (ALL SCRIPTS) -----------------------------------
    
    directory = '/home/ubuntu/Semester10/NLP/Project/scripts'
    
    for filename in os.listdir(directory):
        
        if (filename.endswith(".txt")):
            
            f = os.path.join(directory, filename)
            #print(f)
  
            script = open(f)
            s = script.read()
            removeNewLine = s.replace("\n", " ")
            removeTab = removeNewLine.replace("\t", "")
            
        # ---------------------------------------------------------------------------
            
        # ----------------------------- Preprocessing -------------------------------
            #Lower Case
            lowerCase = removeTab.lower()
            appendApostrophe = lowerCase.replace(" '", "'")
            tokenized = tokenize(appendApostrophe)
            
            #Remove Stop Words
            noMoreStopWords = removeALLStopWords(tokenized)

            #Remove Punctuation
            punctRemoved = removePunct(noMoreStopWords)

            #Remove Extra Spaces
            noMoreExtraSpaces = removeExtraSpaces(punctRemoved)

            #Tokenize The Space Filtered Text
            noMoreExtraSpacesTokenized = tokenize(noMoreExtraSpaces)
            
        # ----------------------------------------------------------------------------
        
        
        # ----------------------------- Feature Extraction ---------------------------
            
            #Feature Extraction
            vector = uniqueWords(tokenize(noMoreExtraSpaces))
        # ----------------------------------------------------------------------------
        
        # ---------------------------- VAD Vectorization -----------------------------
            #Generate VAD Dictionary
            VADDic = genVADDic()

            #Match Script to VAD
            v, a, d = matchScriptToVAD(vector, VADDic)

            #Averaging using window of size 500
            avV,avA, avD = averaging(v, a, d)
            
        # ----------------------------------------------------------------------------
        
        # --------------------------------- Output -----------------------------------  
            #Plot Figure
            plotVAD(avV, avA, avD, f)
            
        # ----------------------------------------------------------------------------
```
